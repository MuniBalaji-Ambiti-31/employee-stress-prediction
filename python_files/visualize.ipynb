```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv('./Dataset/train.csv')
data.describe()
```

```python
data = data.dropna()
data = data.drop(['Employee ID','Date of Joining'],axis=1,inplace=False)
data['Gender'] = data['Gender'].map({'Male' : 1,'Female' :0})
data['WFH Setup Available'] = data['WFH Setup Available'].map({'Yes':1,'No' : 0})
data['Company Type'] = data['Company Type'].map({'Product':1,'Service' : 0})
data
```

```python
sns.set(style='whitegrid')
ax = sns.barplot(x='Burn Rate',y='Gender',data=data)
ax.set_title('Gender vs Burn Rate')
ax.set_xlabel('Burn Rate')
ax.set_ylabel('Gender')
```

```python
filtered_df = data[(data['Mental Fatigue Score'] >= 0) & (data['Mental Fatigue Score'] <= 10) & (data['Designation'] >= 0) & (data['Designation'] <= 5)]

# plot histogram
g = sns.histplot(data=filtered_df, x='Mental Fatigue Score', hue='Designation', multiple='stack', kde=False, bins=10)
g.set(xlabel='Mental Fatigue Score', ylabel='Number of Instances')
plt.show()
```

```python
sns.countplot(data=filtered_df, x='Mental Fatigue Score', hue='Gender')
```

```python
fig = px.bar(data, x='Gender', y='Average Fatigue Score', color='Gender')
fig.show()
```
